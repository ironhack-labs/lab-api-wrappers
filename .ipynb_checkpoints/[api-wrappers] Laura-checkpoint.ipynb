# Function for Playlists

```python
def playlists_data(sp, playlist_id):
    data = []
    results = sp.playlist_items(playlist_id)

    while results:
        tracks = results['items']
        for item in tracks:
            title = item['track']['name']
            artist_names = [artist['name'] for artist in item['track']['artists']]
            artists_string = ", ".join(artist_names)  # Combine all artist names into a single string
            album = item['track']['album']['name']
            album_release = item['track']['album']['release_date']
                
            data.append({
                'Title': title,
                'Artists': artists_string,
                'Album': album,
                'Album Release Date': album_release
                })
        results = sp.next(results)
    
    return pd.DataFrame(data)
```

```python
playlist1 = playlists_data(sp, '1y2s8NZO44F5p1obqnQns3')
playlist2 = playlists_data(sp, '37i9dQZF1DX8AliSIsGeKd')
playlist3 = playlists_data(sp, '51y4OjkmsYww3CQ3uPwzqk')
playlist4 = playlists_data(sp, '5S8SJdl1BDc0ugpkEvFsIL')

# Concatenate the two DataFrames
df_playlists = pd.concat([playlist1, playlist2, playlist3, playlist4], axis = 0, ignore_index=True)
df_playlists
```

```python
df_playlists = df_playlists.drop_duplicates()
df_playlists
```

# Function for a "search"

```python
def search_data(sp, query):
    data = []
    results = sp.search(q = query, type='track')

    while results:
        tracks = results['tracks']['items']
        for item in tracks:
            title = item['name']
            artist_names = [artist['name'] for artist in item['artists']]
            artists_string = ", ".join(artist_names)  # Combine all artist names into a single string
            album = item['album']['name']
            album_release = item['album']['release_date']
                
            data.append({
                'Title': title,
                'Artists': artists_string,
                'Album': album,
                'Album Release Date': album_release
            })
        if results['tracks']['next']:
            results = sp.next(results['tracks'])
        else:
            break
    
    return pd.DataFrame(data)
```

```python
search1 = search_data(sp, 'love')
search2 = search_data(sp, 'sex')
search4 = search_data(sp, 'iron')
search5 = search_data(sp, 'hack')
search6 = search_data(sp, 'portugal')

df_search = pd.concat([search1, search2, search3,search4, search5, search6], axis = 0, ignore_index=True)
df_search
```

```python
df_search = df_search.drop_duplicates()
```

```python
df_final = pd.concat([df_playlists, df_search], axis = 0, ignore_index=True)
df_final
```
