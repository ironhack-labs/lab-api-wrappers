#### Building the function

```python
def extract_playlist_tracks(sp, user, playlist_id):
    running_1 = []
    offset = 0
    limit = 100
    
    playlist_tracks = sp.user_playlist_tracks(user=user, playlist_id=playlist_id, limit=limit, offset=offset)

    while playlist_tracks['items']:

        for item in playlist_tracks['items']:
            artist_name = item['track']['album']['artists'][0]['name']
            disc_name = item['track']['album']['name']
            release_date = item['track']['album']['release_date']
            total_tracks = item['track']['album']['total_tracks']
            duration_ms = item['track']['duration_ms']
            duration_minutes = duration_ms / 60000  # Converting milliseconds to minutes
            explicit = item['track']['explicit']
            song_name = item['track']['name']
            popularity = item['track']['popularity']
            track_number = item['track']['track_number']
            
            running_1.append([artist_name, disc_name, release_date, total_tracks, duration_minutes, explicit, song_name, popularity, track_number])

        # Updating the offset for the next batch of tracks
        offset += limit
        # Geting the next batch of playlist tracks with the updated offset
        playlist_tracks = sp.user_playlist_tracks(user=user, playlist_id=playlist_id, limit=limit, offset=offset)

    # Createing a DataFrame with the extracted data
    running_1_df = pd.DataFrame(running_1, columns=['Artist', 'Disc Name', 'Release Date', 'Total Tracks', 'Duration (Minutes)', 'Explicit', 'Song Name', 'Popularity', 'Track Number'])
    
    return running_1_df

# Calling the function to extract playlist tracks and display the DataFrame
running_df = extract_playlist_tracks(sp, user='1175752327', playlist_id='0oMVDQSeiflEnweDy9ZTN0')
display(running_df)
```
