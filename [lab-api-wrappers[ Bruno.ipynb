```python
def playlists(sp, playlist_id):
    songs = []
    offset = 0
    limit = 100

    while True:
        playlist = sp.playlist_tracks(playlist_id, offset=offset, limit=limit)

        for item in playlist['items']:
            track = item['track']
            title = track['name']
            artist = [artist['name'] for artist in track['artists']]
            album = track['album']['name']
            album_release = track['album']['release_date']

            songs.append({
                'Title': title,
                'Artists': artist,
                'Album': album,
                'Album Release Date': album_release
            })

        offset += limit
        if offset >= playlist['total']:
            break

    return pd.DataFrame(songs)
```
