```python
def getSongIds(playlist_id):
    song_ids = []
    playlist = sp.playlist(playlist_id)
    total_songs = playlist['tracks']['total']
    offset = 0
    while len(song_ids) < total_songs:
        playlist_items = sp.playlist_items(playlist_id, offset=offset)['items']
        for item in playlist_items:
            song_id = item['track']['id']
            if song_id not in song_ids:
                song_ids.append(song_id)
        offset += 100
    return song_ids

getSongIds("spotify:playlist:37i9dQZF1DWXRqgorJj26U")
```

```python
def getArtistIds(playlist_id):
    artists_ids = []    
    for song_id in songs_ids:
        song_info = sp.track(song_id)

        for artist in song_info['artists']:
            if artist['id'] not in artists_ids:
                artists_ids.append(artist['id'])
    return artists_ids
                
artists_ids=getArtistIds("spotify:playlist:37i9dQZF1DWXRqgorJj26U")
artists_ids
```

```python
def getAlbumIds(artist_ids):
    album_ids = []
    for artist_id in artist_ids:
        albums = sp.artist_albums(artist_id, album_type='album')
        for album in albums['items']:
            if album['id'] not in album_ids:
                album_ids.append(album['id'])
    return album_ids
                
album_ids=getAlbumIds(artists_ids)
album_ids
```

```python
def getAllSongs(album_ids):
    song_ids = []
    for album_id in album_ids:
        tracks = sp.album_tracks(album_id)
        for track in tracks['items']:
            if track['id'] not in song_ids:
                song_ids.append(track['id'])
    return song_ids

allsongs_ids=getAllSongs(album_ids)
allsongs_ids
```
